```python
gf = open('/Users/rlynch/Downloads/google.csv')
cf = open('/Users/rlynch/Downloads/covid.csv')
```

```python
import csv
import numpy as np
d_google = {}
with open('/Users/rlynch/Downloads/google.csv', 'r') as gf:
    r = csv.reader(gf)
    data = [[] for i in range(0,50)]
    
    for row in r:
        if 'SUM' in row[0]:
            continue
        if row[0]=='date':
            states = row[1:]
        else:
            for i in range(0,50):
                data[i].append(int(float(row[i + 1])))
    for i in range(0,50):
        d_google[states[i]] = data[i]
    
d_covid = {}
with open('/Users/rlynch/Downloads/covid.csv', 'r') as cf:
    r = csv.reader(cf)
    data = [[] for i in range(0,50)]
    
    for row in r:
        if 'SUM' in row[0]:
            continue
        if row[0]=='date':
            states = row[1:]
        else:
            for i in range(0,50):
                data[i].append(int(float(row[i + 1])))
    for i in range(0,50):
        d = np.convolve(data[i], np.ones((2))/2)
        d_covid[states[i]] = d
    print(d_covid)

            
```
